```python
# -------------------------------------------------------
#                 IMPORTAR LIBRERIAS 
# ------------------------------------------------------

# importar librerias de manipulacion de Datos
import pandas as pd 
import numpy as np
import sys

file = '/mnt/c/DataScience/Projects_Python/Clinica_Prime/clientes_work.xlsx'
excel_file= pd.ExcelFile(file)
meses = excel_file.sheet_names


df_final = pd.DataFrame()

# Agiizar
# Nombre legibles no Deseados Se agregan a una lista llamada `eliminar_columna`
eliminar_columna = [ 'costo', #Redundante con el Total, no hace Referencia a Costo del Material
                     'fotos ',
                     'hora',
                     'hora.1',
                     'motivo de consulta',
                     'metodo de pago',
                     'metodo de pago.1',
                     'anuncio',
                     'deuda']

columnas_reemplazar =  ['tratamiento/procedimiento',
                       'procedimiento',
                       'notas.1']


map_columnas = {'tratamiento/procedimiento': 'tratamiento',
                       'procedimiento':'tratamiento',
                       'notas.1':'notas'}

nombres_base = ['edad',
                'nombre',
                'tel√©fono',
                'sexo',
                'dni',
                'nacimiento',
                'distrito',
                'total',
                'notas',
                'pp',
                'tratamiento',  
                'fecha']    

for mes in meses:
    df = excel_file.parse(mes)
    
    renombradas_encontradas = []
    descartados_encontrados = []
    desconocidas_encontradas = []
   
    
    try: 
        df.columns = [str(col).lower() for col in df.columns]
        for col in df.columns:
            if col not in nombres_base:
                # Eliminar Columnas no Deseadas
                if col in eliminar_columna:
                    print(f"\n Se Encontro columna a eliminar {col}")
                    df.drop(col, axis=1, inplace=True)
                    descartados_encontrados.append(col)
                    
                # Renombrar columnas Conocidas segun el Mapeo
                df.rename(columns=map_columnas, inplace=True)
                for col_map in map_columnas.keys():
                    if col_map in df.columns:
                        print(f"Renombrando Columna Conocida {col_map} -> {map_columnas[col_map]}")
                        renombradas_encontradas.append(col_map)
                
                # Identificar columnas desconocidas por el Tipo de Datos
                if col not in eliminar_columna and col not in columnas_reemplazar:
                    dtype = df[col].dtype
                    print(f"\nüîç Analizando columna desconocida: {col} (tipo: {dtype})")
                    if dtype == 'datetime64[ns]':
                        df.rename(columns={col: 'fecha'}, inplace=True)
                    elif dtype == 'object':
                        df.rename(columns={col: 'tratamient'}, inplace=True)
        print(df.columns)
        
        df_final = pd.concat([df_final, df], ignore_index = True)

    except Exception as e:
        print(f"‚ùå MISI√ìN ABORTADA en FASE 1. Error: {e}")
        
        
        

```

```python
# ======================================================================
# üèõÔ∏è CIRUG√çA RECONSTRUCTIVA, PASO 1: FUSI√ìN DE GEMELOS
# ======================================================================
# PRERREQUISITO: El DataFrame `df_final` existe.

print("--- Iniciando Fusi√≥n de Columnas Duplicadas ---")
print(f"Columnas antes de la fusi√≥n: {df_final.columns.tolist()}")

# La Anatom√≠a: `.fillna()`
# El Prop√≥sito del Artesano: Le decimos a Pandas: "En la columna 'tratamiento',
# si encuentras una celda vac√≠a (NaN), intenta rellenarla con el valor
# de la misma fila pero de la columna 'tratamient'".
# Esto es un trasplante de datos.
df_final['tratamiento'] = df_final['tratamiento'].fillna(df_final['tratamient'])

# La Anatom√≠a: `.drop()`
# El Prop√≥sito del Artesano: Ahora que hemos salvado los datos de la columna
# redundante, la destruimos.
df_final.drop(columns=['tratamient'], inplace=True)

print("\n‚úÖ Fusi√≥n completada.")
print(f"Columnas despu√©s de la fusi√≥n: {df_final.columns.tolist()}")

print("\n--- Verificaci√≥n de la Estructura Post-Fusi√≥n ---")
print(df_final.info())
```

```python
from limpieza_utils import convertir_a_fechas, limpiar_y_convertir_a_numerico

print("\n--- Iniciando la Cirug√≠a de Tipos de Datos ---")
print("\n--- Estado PRE-Operatorio del Paciente ---")
print(df_final.info())
```

```python

# --- Aplicando las Herramientas Quir√∫rgicas ---

# 1. Operaci√≥n de Fechas
df_final = df_final.pipe(convertir_a_fechas, nombre_columna='fecha')
df_final = df_final.pipe(convertir_a_fechas, nombre_columna='nacimiento') # Tambi√©n para nacimiento

# 2. Operaci√≥n de Columnas Num√©ricas
columnas_a_limpiar_numeric = ['total', 'dni', 'tel√©fono', 'edad']
for col in columnas_a_limpiar_numeric:
    df_final = df_final.pipe(limpiar_y_convertir_a_numerico, nombre_columna=col)
    
    # 2.1 Post-Operatorio: Conversi√≥n a Entero si es posible
    # Despu√©s de limpiar, `dni`, `tel√©fono`, `edad` son `float64` por los `NaN`.
    # Los convertimos a `Int64` para un formato m√°s limpio.
    if col in ['dni', 'tel√©fono', 'edad']:
        # `errors='ignore'` es por si la columna no existiera, aunque ya lo comprobamos.
        df_final[col] = df_final[col].astype('Int64', errors='ignore')

```
