```python
# -------------------------------------------------------
#                 IMPORTAR LIBRERIAS 
# ------------------------------------------------------

# importar librerias de manipulacion de Datos
import pandas as pd 
import numpy as np
import sys

file = '/mnt/c/DataScience/Projects_Python/Clinica_Prime/data/clientes_work.xlsx'
excel_file= pd.ExcelFile(file)
meses = excel_file.sheet_names


df_final = pd.DataFrame()

# Agiizar
# Nombre legibles no Deseados Se agregan a una lista llamada `eliminar_columna`
eliminar_columna = [ 'costo', #Redundante con el Total, no hace Referencia a Costo del Material
                     'fotos ',
                     'hora',
                     'hora.1',
                     'motivo de consulta',
                     'metodo de pago',
                     'metodo de pago.1',
                     'anuncio',
                     'deuda']

columnas_reemplazar =  ['tratamiento/procedimiento',
                       'procedimiento',
                       'notas.1']


map_columnas = {'tratamiento/procedimiento': 'tratamiento',
                       'procedimiento':'tratamiento',
                       'notas.1':'notas'}

nombres_base = ['edad',
                'nombre',
                'tel√©fono',
                'sexo',
                'dni',
                'nacimiento',
                'distrito',
                'total',
                'notas',
                'pp',
                'tratamiento',  
                'fecha']    

for mes in meses:
    df = excel_file.parse(mes)
    
    renombradas_encontradas = []
    descartados_encontrados = []
    desconocidas_encontradas = []
   
    
    try: 
        df.columns = [str(col).lower() for col in df.columns]
        for col in df.columns:
            if col not in nombres_base:
                # Eliminar Columnas no Deseadas
                if col in eliminar_columna:
                    print(f"\n Se Encontro columna a eliminar {col}")
                    df.drop(col, axis=1, inplace=True)
                    descartados_encontrados.append(col)
                    
                # Renombrar columnas Conocidas segun el Mapeo
                df.rename(columns=map_columnas, inplace=True)
                for col_map in map_columnas.keys():
                    if col_map in df.columns:
                        print(f"Renombrando Columna Conocida {col_map} -> {map_columnas[col_map]}")
                        renombradas_encontradas.append(col_map)
                
                # Identificar columnas desconocidas por el Tipo de Datos
                if col not in eliminar_columna and col not in columnas_reemplazar:
                    dtype = df[col].dtype
                    print(f"\nüîç Analizando columna desconocida: {col} (tipo: {dtype})")
                    if dtype == 'datetime64[ns]':
                        df.rename(columns={col: 'fecha'}, inplace=True)
                    elif dtype == 'object':
                        df.rename(columns={col: 'tratamient'}, inplace=True)
        print(df.columns)
        
        df_final = pd.concat([df_final, df], ignore_index = True)

    except Exception as e:
        print(f"‚ùå MISI√ìN ABORTADA en FASE 1. Error: {e}")
        
        
        

```

```python
#================================================================================
# BRANCH - ETL - SQL
#=================================================================================

from sqlalchemy import create_engine, text 
import sys

from src.hidden import secret_credentials
secrets = secret_credentials()

user = secrets['user']
password = secrets['password']
host = secrets['host']
port = secrets['port']
database = secrets['database']


# =/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=
#  LA FUNCI√ìN ORQUESTADORA - EN TU NOTEBOOK DE PYTHON
# =/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=/=
import traceback
from pathlib import Path
from sqlalchemy import text # ¬°Importante importar `text`!

def run_sql_file(engine, sql_path, verify_query=None, verify_label=""):
    """
    Lee un archivo .sql y lo ejecuta contra la base de datos de la forma m√°s robusta.
    """
    path = Path(sql_path)
    if not path.is_file():
        print(f"‚ùå Error: El archivo de plano '{path.name}' no fue encontrado en la ruta.")
        return

    print(f"\n  ----- Running SQL: {path} ------")
    
    try:
        # --- PASO 1: LEER EL CONTENIDO DEL ARCHIVO ---
        # Leemos el archivo .sql completo como una √∫nica cadena de texto.
        with open(path, 'r', encoding='utf-8') as f:
            sql_script = f.read()
        
        # --- PASO 2: EJECUCI√ìN DIRECTA Y ROBUSTA ---
        # La Anatom√≠a: `with engine.connect() as connection:`
        # El Prop√≥sito: Abrimos una conexi√≥n directa y la gestionamos de forma segura.
        # La Anatom√≠a: `connection.execute(text(sql_script))`
        # El Prop√≥sito: `text()` le dice a SQLAlchemy "trata esta cadena de texto
        # como un bloque de SQL puro y ejec√∫talo tal cual, sin intentar ser demasiado listo".
        # Esto evita las capas internas que causaban el TypeError.
        # `connection.commit()` asegura que los cambios se guarden permanentemente.
        with engine.connect() as connection:
            connection.execute(text(sql_script))
            connection.commit() # Expl√≠citamente guardamos la transacci√≥n.

        print(f"   Success: {path.name} executed")

        # --- PASO 3: VERIFICACI√ìN (Sin cambios) ---
        if verify_query:
            with engine.connect() as connection:
                result = connection.execute(text(verify_query)).scalar()
                print(f"   Verification -> {verify_label}: {result} registros.")

    except Exception as e:
        print(f"   Failed Executing: {path}")
        print(f"   Script head: {sql_script[:200]}...")
        print("   Traceback:")
        traceback.print_exc()
        raise e
    
try: 
    engine = create_engine(f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{database}")
except Exception as e: 
    print(f"Fallo de Conexion: {e}")
    sys.exit() 
# Creacion de la Funcion para ejecutar los Archivos `sql`
'''
def run_sql_file(engine, sql_path: str, verify_query: str | None=None, verify_label:str = "verification"):

    path = Path(sql_path)
    print(f"\n  ----- Running SQL: {path} ------")
    if not path.exists():
        raise FileNotFoundError(f"SQL file not found: {path}")
    
    try: 
        sql = path.read_text(encoding="utf-8")
        # exec_driver _sql maneja sripts con multiples sentencias separadas por ';'
        with engine.begin() as conn:
            conn.exec_driver_sql(sql)
        print(f"Success: {path.name} executed")
        
        if verify_query: 
            with engine.connect() as conn:
                df_check = pd.read_sql_query(verify_query, conn)
                print(f"{verify_label}: \n{df_check.to_string(index=False)}")
    
    except Exception as e:
        print(f"Failed Executing: {path}")
        snippet = sql[:300].replace("\n", " ")
        print(f"   Script head: {snippet} ...")
        print("   Traceback:")
        traceback.print_exc()
        raise e  # re-lanza para que cualquier pipeline/CI lo detecte

            
                
                
'''
```
